# Plot a bar graph of male vs female buyers participated in the sales

```python
no=df.groupby("Buyer Gender").size().reset_index(name="count")
no
```

```python
plt.figure(figsize=(5,10))
plt.bar(no["Buyer Gender"],no["count"],color=["orange","green"])
plt.xlabel("Gender")
plt.ylabel("Number of Buyers")
plt.title("Male vs Female buyers")
plt.show()
```

Buyer Female :5052 , 
Male :4948

# Find the top 5 cars based on their sales price

```python
top=df.nlargest(5,"Sale Price")
top
```

```python
top=top.loc[:,["Make","Model","Nickname","Sale Price"]]
top
```

```python
top["car"]=top["Make"]+" "+top["Model"]
top
```

```python
plt.figure(figsize=(10,8))
plt.bar(top["car"],top["Sale Price"],color=["red","orange","green","blue","yellow"])
plt.xlabel("Car Model")
plt.ylabel("Price")
plt.title("Top 5 cars based on their sales price")
plt.show()
```

Top 5 cars based on their sales price are Volkswagen Jetta III, Audi A6, Lexus LX, Mazda MX-6, Lexus ES.
